## Querying GitHub

```python
from github import Github

ACCESS_TOKEN = 'MY_ACCESS_TOKEN'
USER = "vuejs"
#REPO = "vue-next"

g1 = Github(ACCESS_TOKEN)
g2 = Github(USER)

user = g2.get_user()
```

```python
#Get repositories for an organization
org = g1.get_organization('vuejs')
repos = org.get_repos()
repos_id_lsit= []
repos_name_lsit = []
for repo in repos:
    repos_id_lsit.append(repo.id)
    repos_name_lsit.append(repo)
```

```python
#Get issues for a repo

def get_issues_for_a_repo(repo):
    
    core_rate_limit = g1.get_rate_limit().core
    #print("get_issues", core_rate_limit)
    
    if core_rate_limit.remaining > 500: 
    
        issues = repo.get_issues(state="closed")
        issues_name_list = []
        for issue in issues:
            issues_name_list.append(issue)
            
    else:  
        reset_timestamp = calendar.timegm(core_rate_limit.reset.timetuple())
        sleep_time = reset_timestamp - calendar.timegm(time.gmtime())+ 5  # add 5 seconds to be sure the rate limit has been reset
        time.sleep(sleep_time)
    
        issues = repo.get_issues(state="all")
        issues_name_list = []
        for issue in issues:
            issues_name_list.append(issue)
    
    return issues_name_list
```

```python
# query the issue tracker, making sure that the rate limit is not exceeded 

def create_issue_dic(issues_name_list): 
    
    issue_dic = {}
    
    for issue in issues_name_list: 
        
        core_rate_limit = g1.get_rate_limit().core
        #print(issue.number, core_rate_limit)
        
        if core_rate_limit.remaining > 500: 

            temp_dict = {}
            temp_dict["title"] = issue.title
            temp_dict["created_at"] = issue.created_at
            temp_dict["created_by_id"] = issue.user.login
            temp_dict["created_by_name"] = issue.user.name
            try:
                user_name = str(issue.user.name).split(" ")
                if user_name[0] in list(names_dict.keys()): 
                    temp_dict["gender"] = names_dict[user_name[0]]
                elif user_name[1] in names_dict.keys(): 
                    temp_dict["gender"] = names_dict[user_name[1]]
                else: 
                    temp_dict["gender"] = "unkown"
            except (IndexError, AttributeError):
                temp_dict["gender"] = "unkown"
            temp_dict["closed_at"] = issue.closed_at
            
            
            temp_dict["comments"] = {}
            for comment in issue.get_comments(): 
                temp_comment_dic = {}
                temp_comment_dic["user"] = comment.user.login
                temp_comment_dic["name"] = comment.user.name
                try:
                    user_name = str(comment.user.name).split(" ")
                    if user_name[0] in list(names_dict.keys()): 
                        temp_comment_dic["gender"] = names_dict[user_name[0]]
                    elif user_name[1] in names_dict.keys(): 
                        temp_comment_dic["gender"] = names_dict[user_name[1]]
                    else: 
                        temp_comment_dic["gender"] = "unkown"
                except (IndexError, AttributeError):
                    temp_comment_dic["gender"] = "unkown"
                temp_comment_dic["created_at"] = comment.created_at
                temp_dict["comments"][comment.id] = temp_comment_dic
                
            issue_dic[issue.number] = temp_dict
            
        else:
            reset_timestamp = calendar.timegm(core_rate_limit.reset.timetuple())
            sleep_time = reset_timestamp - calendar.timegm(time.gmtime()) + 10  # add 10 seconds to be sure the rate limit has been reset
            time.sleep(sleep_time)
            
            temp_dict = {}
            temp_dict["title"] = issue.title
            temp_dict["created_at"] = issue.created_at
            temp_dict["created_by_id"] = issue.user.login
            temp_dict["created_by_name"] = issue.user.name
            try:
                user_name = str(issue.user.name).split(" ")
                if user_name[0] in list(names_dict.keys()): 
                    temp_dict["gender"] = names_dict[user_name[0]]
                elif user_name[1] in names_dict.keys(): 
                    temp_dict["gender"] = names_dict[user_name[1]]
                else: 
                    temp_dict["gender"] = "unkown"
            except (IndexError, AttributeError):
                temp_dict["gender"] = "unkown"
            temp_dict["closed_at"] = issue.closed_at
            
            
            temp_dict["comments"] = {}
            for comment in issue.get_comments(): 
                temp_comment_dic = {}
                temp_comment_dic["user"] = comment.user.login
                temp_comment_dic["name"] = comment.user.name
                try:
                    user_name = str(comment.user.name).split(" ")
                    if user_name[0] in list(names_dict.keys()): 
                        temp_comment_dic["gender"] = names_dict[user_name[0]]
                    elif user_name[1] in names_dict.keys(): 
                        temp_comment_dic["gender"] = names_dict[user_name[1]]
                    else: 
                        temp_comment_dic["gender"] = "unkown"
                except (IndexError, AttributeError):
                    temp_comment_dic["gender"] = "unkown"
                temp_comment_dic["created_at"] = comment.created_at
                temp_dict["comments"][comment.id] = temp_comment_dic
                
            issue_dic[issue.number] = temp_dict
        
    return issue_dic
```

```python
# creation of a time encoder to be able to store the data in .txt formate
# code is based on https://stackoverflow.com/questions/44630103/how-to-write-and-read-datetime-dictionaries

class DateTimeAwareEncoder(json.JSONEncoder):
    def default(self, o):
        if isinstance(o, datetime):
            return o.isoformat()

        return json.JSONEncoder.default(self, o)
```

```python
# write the scraped data in a .txt file 

repo_dict = {}
for repo in repos: 
    core_rate_limit = g1.get_rate_limit().core
    print(repo.name, core_rate_limit)
    
    issue_dict = {}
    issues_name_list = get_issues_for_a_repo(repo)
    issue_dict = create_issue_dic(issues_name_list)
    repo_dict[repo.name] = issue_dict
    
    with open('info_repo_and_issues.txt', 'w') as outfile:
        json.dump(repo_dict, outfile, cls=DateTimeAwareEncoder)
    outfile.close()
```
